```python
import sage.graphs.strongly_regular_db as srg
from sage.misc.unknown import UnknownClass
```

```python
potential_parameters = srg.apparently_feasible_parameters(3000)
real_parameters = set()
unknown_parameters = set()
print(f"{len(potential_parameters)} potentially feasible parameter sets")
for v, k, lambda_param, mu in potential_parameters:
    result = srg.strongly_regular_graph(v, k, lambda_param, mu, existence=True)
    if type(result) != UnknownClass and result:
        real_parameters.add((v, k, lambda_param, mu))
    else:
        unknown_parameters.add((v, k, lambda_param, mu))

print(f"{len(real_parameters)} real parameter sets")
print(f"{len(unknown_parameters)} unknown parameter sets")

for v, k, lambda_param, mu in real_parameters:
    print(f"{v}, {k}, {lambda_param}, {mu}")
```
